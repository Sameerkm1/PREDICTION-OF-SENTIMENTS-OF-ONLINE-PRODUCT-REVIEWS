```python
# Read the csv file
data = pd.read_csv ('/content/drive/MyDrive/amazon_lemmat_bal_drop_3_dataset.csv')
#df = pd.DataFrame(data, columns= ['Reviews']) #to read review colomn only
# To get the first n rows
data.head(10)
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Label'].value_counts()
```

```python
## CountVector and RF
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 20000, ngram_range=(1,1))
```

```python
#tfidf =  TfidfVectorizer(max_features = 20000, ngram_range=(1,3),analyzer='char')
```

```python
#data['Reviews_lemmatizer'] = data['Reviews_lemmatizer'].repr()
#x = tfidf.fit_transform(data['Reviews_lemmatizer'])
#y= data['Label']
x = cv.fit_transform(data['Reviews_lemmatizer'])
y = data['Label']
```

```python
x.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=0)
```

```python
print(x_train)
```

```python
print(y_train)
```

```python
### Random Forest 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score

rf = RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred =rf.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred, digits=4))
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score 
cm = confusion_matrix(y_test, y_pred)
print ("Accuracy : ", accuracy_score(y_test, y_pred))
cm
```
