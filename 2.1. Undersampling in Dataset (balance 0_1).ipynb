```python
Labels = 'Label'
```

```python
minority_class_len = len(data[data[Labels] == 0])
print(minority_class_len)
```

```python
majority_class_indices = data[data[Labels] == 1].index
print(majority_class_indices)
```

```python
#Undersampling the samples
random_majority_indices = np.random.choice(majority_class_indices,minority_class_len,replace=False)
print(len(random_majority_indices))
```

```python
minority_class_indices = data[data[Labels] == 0].index
print(minority_class_indices)
```

```python
#Concatenating both indices: obtaining a new string that contains both of the original strings
under_sample_indices = np.concatenate([minority_class_indices, random_majority_indices])
```

```python
#creating new dataframe
under_sample = data.loc[under_sample_indices]
under_sample.head()
```

```python
under_sample_bal = under_sample.drop(['Unnamed: 0','Unnamed: 0.1'], axis=1)
under_sample_bal.head()
```
